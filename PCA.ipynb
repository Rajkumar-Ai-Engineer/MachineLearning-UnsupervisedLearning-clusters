# **Libs**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

# **Load Iris**

```python
dataset = load_iris()
```

# **Segregating the data**

```python
x = dataset.data
y = dataset.target
names =dataset.target_names
```

# **PCA**

```python
from sklearn.decomposition import PCA

model = PCA(n_components=2)

y_means = model.fit(x).transform(x)
```

```python
plt.figure()
colors = ['deeppink','lime','skyblue']

for color,i,target_names in zip(colors,[0,1,2],names):
    plt.scatter(y_means[y==i,0],y_means[y==i,1],c=color,label=target_names,marker='*',lw=2)
plt.title("IRIS CLUSTERING")
plt.legend()
plt.show()

```
